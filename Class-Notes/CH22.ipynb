```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
plants = ["Lemon", "Lime", "Orange"]
fruit_yield = [89, 43, 12]

plt.bar(plants, fruit_yield)
```

```python
data = [["Lemon", 89], ["Lime", 43], ["Orange", 12]] #creates a dataframe
citrus = pd.DataFrame(data, columns=["plant","fruit_yield"]) #assigns column labels to create a dataframe

sns.barplot(citrus, x="fruit_yield", y="plant", orient="y") #creates the visualization, and orient="y" specifies a bar chart rather than column
```

```python
# matplotlib method
butterflies=[12,34,89]
pollinator_participants=[10,56,102]

plt.scatter(pollinator_participants, butterflies)
```

```python
# seaborn method
data = [[12,10], [34,56], [89,102]]
pollinator_program = pd.DataFrame(data, columns=["butterflies","pollinator_participants"])

sns.scatterplot(data=pollinator_program, x="pollinator_participants", y="butterflies")
```
