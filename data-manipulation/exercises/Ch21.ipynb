```python
titanic["survived_reformatted"] = titanic["survived"].replace({0 : False, 1: True}) #the left side of the equal sign assigned the name to the new column within
#the datafram. The right side of the equal sign established what should be in that new column. It's replacing integers with boolean
titanic.head(20)
```

```python
def under_age_21_survivors(titanic): #defines a new funtion with that name on the dataframe as the range of data
    age = titanic['age'] #assigs the column
    alive = titanic['alive'] #assigns the column

    if age <= 20 and alive == "yes": #applies the conditions
        return True 
    else:
        return False

titanic["under_21_survivors"] = titanic.apply(under_age_21_survivors, axis=1) #creates a new column with the title of the new function and applies the conditions
print(titanic["under_21_survivors"].value_counts()) #prints the new aggregate data based on the values count only. 
```

```python
# syntax for concatenate
# dataset_one = [1, 2, 3]
# dataset_two = ["four", "five", "six"]

# dataframe_one = pd.DataFrame(dataset_one)
# dataframe_two = pd.DataFrame(dataset_two)

# combined_dataframe = pd.concat([dataframe_one, dataframe_two])

movie_dictionary = {
    'Name': ["Interstellar", "Pride and Prejudice", "Inception", "Barbie"], 
    'Release': [2014, 2005, 2010, 2003], 
}

genre_rating_dictionary = {
    'Genre': ["Science Fiction", "Novel", "Science Fiction", "Comedy"], 
    'Rotten Tomatoes Score': [73, 87, 87, 88]
}

movie_dataframe = pd.DataFrame(movie_dictionary)
genre_rating_dataframe = pd.DataFrame(genre_rating_dictionary)

movie_genre_dataframe = pd.concat([movie_dataframe, genre_rating_dataframe], axis=1)
print(movie_genre_dataframe)
```

```python
#adding a series to an existing dataframe
data = ["movies", "genres", "release_year"]
series_data = ["interstellar", "pride and prejudice", "barbie"]

example_dataframe = pd.DataFrame(data)
example_series = pd.Series(series_data)

concat_series_dataframe = pd.concat([example_dataframe, example_series], axis=1)
print(concat_series_dataframe)
```
